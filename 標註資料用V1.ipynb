```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
```

```python
import os, json 
import pandas as pd 
#resourcedir = '臺灣新北地方法院_改_V2'
resourcedir = 'TEST'

path_to_json = 'J:\Legal_Judgment\\' + resourcedir
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')] 
print(json_files[0])

bolRegex = 0
dir = 'dataR15'
```

```python
with open('J://Legal_Judgment//'+ resourcedir + '//'+json_files[0],"r",encoding="utf-8") as f:
    dt = json.load(f)
```

```python
dic = {
    "id": 0,
    "court": [dt["court"]], 
    "date": [dt["date"]],
    "no": [dt["no"]],
    "sys": [dt["sys"]],
    "reason": [dt["reason"]],
    "type": [dt["type"]],
    "mainText": [dt["mainText"]],
    "judgement": [dt["judgement"]],
    #"judgement_SIT": [dt["judgement_SIT"]],
    "Situation": [dt["Situation"]]
}
df = pd.DataFrame(dic)
```

```python
for x in range(1,len(json_files)):
    with open('J://Legal_Judgment//' + resourcedir + '//'+json_files[x],"r",encoding="utf-8") as f:
        dt = json.load(f)
    #ltmp = [225,1156,1215,1056,22,33,235,676,778]
    #if (x in ltmp):
    #    continue

    try:
        if (str(dt["judgement"]).index(str(dt["Situation"])) < 0):
            continue
        if (len(tokenizer.tokenize(str(dt["Situation"]))) > 512):
        #if (len(str(dt["Situation"])) > 512):
            continue

        new = pd.DataFrame({
        "id" : x,
        "court": [dt["court"]],
        "date": [dt["date"]],
        "no": [dt["no"]],
        "sys": [dt["sys"]],
        "reason": [dt["reason"]],
        "type": [dt["type"]],
        "mainText": [dt["mainText"]],
        "judgement": [dt["judgement"]],
        #"judgement_SIT": [dt["judgement_SIT"]],
        "Situation": [dt["Situation"]]
        })

        df = df.append(new)

    except:
        print(x)
```

```python
class Data(object):
    def __init__(self):
        self.paragraphs = []
        self.id = "0"
        self.title = ""

class Djson(object):
    def __init__(self):
        self.version = "v1.0"
        self.data = []
 
```

```python
djson2 = Djson()
for index, row in test.iterrows():
    try:
        d = Data()
        d.paragraphs = [{"id" : "TEST_" + str(row.id), 
                      #"context" : row.reason + "[SEP]" + row.judgement, 
                      "context" : row.judgement, 
                      "qas":[{
                          "question" : "?",
                          "id" : "TEST_" + str(row.id) + "_QUERY_0",
                          "answers" :[{
                              "text" : row.Situation,
                              "answer_start" : row.judgement.index(row.Situation) #+ len(row.reason + "[SET]")
                          }]
                      }]
                    }
                    ]
        d.id = "TEST_" + str(row.id)
        d.title = row.no
        djson2.data.append(d)
    except:
        print(index)
        print(row.no)
```

```python
y = json.dumps(djson2, default=lambda o: o.__dict__, ensure_ascii=False)
```

```python
with open('./QA_Test1/HBert/' + dir + '/Test_500.json', 'w', encoding='utf8') as outfile:
    json.dump(djson2, outfile, default=lambda o: o.__dict__ , ensure_ascii=False)
    
```
